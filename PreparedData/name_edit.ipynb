```python
def name_edit(count,path,idlist):
    filelist = os.listdir(path)
    for files in filelist:
            print(files)
            if count >97:
                newname=files[0:-6]+str(idlist[count])+".nii"
                print(newname)
                olddir=os.path.join(path,files)
                os.renames(olddir,os.path.join(path,newname));
                count = count +1
            else:
                count = count +1
```
